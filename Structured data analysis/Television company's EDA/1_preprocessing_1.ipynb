```python
# I import pandas
import pandas as pd

# This reads the data
df_raw_content = pd.read_pickle('banijay_op1data_content_raw.pkl')

# I create the lists for the new columns
date_time_start = []
date_time_end = []

# I make a for loop to add together date and time, without the milliseconds
for date, time in zip (df_raw_content['date'], df_raw_content['start']):
    date_time_start.append(date + ' ' + time [:-3])

# I create the column to which I add the combined data, turned into the right format 
df_raw_content['date_time_start'] = pd.to_datetime(date_time_start, format= '%d-%m-%Y %H:%M:%S')

# I make a for loop to add together date and time, without the milliseconds again
for date, time in zip (df_raw_content['date'], df_raw_content['end']):
    date_time_end.append(date + ' ' + time [:-3])

# I create the column to which I add the combined data, turned into the right format again
df_raw_content['date_time_end'] = pd.to_datetime(date_time_end, format= '%d-%m-%Y %H:%M:%S')

# I print the new columns
df_raw_content.to_csv('banijay_op1data_content_processed.csv', compression='zip')
```

```python
# I create lists to append later
split_id = []
fragment = []
show_id = []

# I split id into three parts, two of which I'm going to use
for id in df_raw_content['id']:
    split_id.append(str.rsplit(id, '_', 2))

# I append show_id and fragment with the parts that they need for each row
for i in range(len(df_raw_content['id'])):
    show_id.append(split_id[i][0])
    fragment.append(split_id[i][1])

# I put these lists into two columns
df_raw_content['show_id'] = show_id
df_raw_content['fragment'] = fragment

# I print the columns
print(df_raw_content['show_id'])
print(df_raw_content['fragment'])

```
