```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#I load the data into a variable
cannabis = pd.read_csv('C:/Users/Dániel/Desktop/Projekt/Cannabis_data_preprocessed.csv')
cannabis.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
cannabis.head(10)
```

```python
cannabis['Effect_1'] = cannabis['Effect_1'].astype(str)
cannabis['Effect_2'] = cannabis['Effect_2'].astype(str)
cannabis['Effect_3'] = cannabis['Effect_3'].astype(str)
cannabis['Effect_4'] = cannabis['Effect_4'].astype(str)
cannabis['Effect_5'] = cannabis['Effect_5'].astype(str)
```

```python
# Get the 'Effect' columns
effect_cols = cannabis.filter(like='Effect').columns.tolist()

# Replace all instances of 'Energetic' in Effect columns with NaN
cannabis[effect_cols] = cannabis[effect_cols].replace('Energetic', np.nan)

# Replace all NaNs in Effect columns with the corresponding values in Effect_1
for col in effect_cols:
    if col != 'Effect_1':
        cannabis[col] = np.where(cannabis[col].isna(), cannabis['Effect_1'], cannabis[col])

# Replace all instances of 'Energetic' in Effect_1 with NaN
cannabis['Effect_1'] = cannabis['Effect_1'].replace('Energetic', np.nan)

# Replace all NaNs in Effect_1 with the corresponding values in the other Effect columns
cannabis['Effect_1'] = np.where(cannabis['Effect_1'].isna(), cannabis[effect_cols].bfill(axis=1)['Effect_1'], cannabis['Effect_1'])

```
