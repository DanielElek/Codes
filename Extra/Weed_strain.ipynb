```python
import pandas as pd
```

```python
#I load the data into a variable
cannabis = pd.read_csv('C:/Users/Dániel/Desktop/Projekt/cannabis.csv')
```

```python
#I check the head 
cannabis.head(10)
```

```python
#I turn the Flavor column into string to be able to split it and to delete 'nan' rows
cannabis['Flavor'] = cannabis['Flavor'].astype(str)
```

```python
#I delete the undesired rows
cannabis = cannabis[cannabis['Effects'] != 'None']
cannabis = cannabis[cannabis['Flavor'] != 'nan']
```

```python
#I check if they were deleted
result = cannabis.loc[cannabis['Effects'] == 'nan']
print(result)
result = cannabis.loc[cannabis['Flavor'] == 'nan']
print(result)
```

```python
#I check the head again
cannabis.head(100)
```

```python
# I create lists to append later
splitter = []
Flavor_1 = []
Flavor_2 = []
Flavor_3 = []

# I split Flavor into three parts
for Flavor in cannabis['Flavor']:
    splitter.append(str.rsplit(Flavor, ',', 3))

# I append Flavor_1, Flavor_2 and Flavor_3
for i in range(len(cannabis['Flavor'])):
    Flavor_1.append(splitter[i][0])
    if len(splitter[i]) > 1:
        Flavor_2.append(splitter[i][1])
    else:
        Flavor_2.append('')
    if len(splitter[i]) > 2:
        Flavor_3.append(splitter[i][2])
    else:
        Flavor_3.append('')

# I put these lists into three columns
cannabis['Flavor_1'] = Flavor_1
cannabis['Flavor_2'] = Flavor_2
cannabis['Flavor_3'] = Flavor_3

# I print the columns
print(cannabis['Flavor_1'].head(20))
print(cannabis['Flavor_2'].head(20))
print(cannabis['Flavor_3'].head(20))
```

```python
#I delete the original Flavor column
cannabis.drop('Flavor', axis=1, inplace=True)
```

```python
# I create lists to append later
splitter = []
Effect_1 = []
Effect_2 = []
Effect_3 = []
Effect_4 = []
Effect_5 = []

# I split Effects into five parts
for Effects in cannabis['Effects']:
    splitter.append(str.rsplit(Effects, ',', 5))

# I append Effect_1, Effect_2, Effect_3, Effect_4 and Effect_5
for i in range(len(cannabis['Effects'])):
    Effect_1.append(splitter[i][0])
    if len(splitter[i]) > 1:
        Effect_2.append(splitter[i][1])
    else:
        Effect_2.append('')
    if len(splitter[i]) > 2:
        Effect_3.append(splitter[i][2])
    else:
        Effect_3.append('')
    if len(splitter[i]) > 3:
        Effect_4.append(splitter[i][3])
    else:
        Effect_4.append('')

    if len(splitter[i]) > 4:
        Effect_5.append(splitter[i][4])
    else:
        Effect_5.append('')
# I put these lists into five columns
cannabis['Effect_1'] = Effect_1
cannabis['Effect_2'] = Effect_2
cannabis['Effect_3'] = Effect_3
cannabis['Effect_4'] = Effect_4
cannabis['Effect_5'] = Effect_5

# I print the columns
print(cannabis['Effect_1'].head(20))
print(cannabis['Effect_2'].head(20))
print(cannabis['Effect_3'].head(20))
```

```python
#I delete the original Effects column
cannabis.drop('Effects', axis=1, inplace=True)
```
