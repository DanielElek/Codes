```python
import json
import pandas as pd

# I read in the json file
with open('banijay_op1data_twitter_raw.json', 'rb') as f:
    twitter_json = json.load(f)

# I convert the json file to a pandas dataframe
df_op1_twitter = pd.json_normalize(twitter_json)

df_op1_twitter.head()
```

```python
# I turn the column into datetime format
pd.to_datetime(df_op1_twitter['created_at'])
```

```python
# I drop the rows that are not needed while creating a new variable for it
df_twitter = df_op1_twitter.dropna(subset=['referenced_tweets'])
```

```python
# I create a new column with the name 'date' on which to merge the two datasets, then I turn it into a string to be able to split it
df_twitter['date'] = df_twitter['created_at']
df_twitter['date'] = df_twitter['date'].astype(str)
print(df_twitter['date'].info)
```

```python
# I split the 'date' column so that I will only have the year, month and day, as this is what the 'date' column in the content_ratings dataset has too.
# Then I turn this into a datetime object
split_date = []
split_date = df_twitter['date'].str.rsplit('T').str[0]
print(split_date.head)
df_twitter['date'] = split_date
df_twitter['date'] = pd.to_datetime(df_twitter['date'])
```

```python
# I import the previous dataset and turn the 'date' column into a datetime object
data = pd.read_csv('banijay_op1data_content_ratings2.csv')
data['date'] = pd.to_datetime(data['date'])
print(data.info)
```

```python
# I merge the two datasets on the 'date' column
merged_data = pd.merge(data, df_twitter, on='date')
print(merged_data.info)
```

```python
# I check if it was successful
print(merged_data.columns)
```

```python
# I export the the new dataset
merged_data.to_csv('banijay_op1data_content_ratings_twitter.csv')
```
